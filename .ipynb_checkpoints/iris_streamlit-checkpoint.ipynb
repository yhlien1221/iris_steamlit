```python
#File -> save as -> type the path where you want to save file. ex: "/work/iris/iris.ipynb"
#Import packages
import os as os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
os.getcwd()
#set up work dict
os.chdir('/home/jovyan/work/iris')
```

```python
#Read iris.csv by pandas
df = pd.read_csv ('iris.csv')
df.head() #list the first five rows of data
df.dtypes #list the types for all columns in data
######note######
#type(df.head)
#<class 'method'>
#type(df.head())
#<class 'pandas.core.frame.DataFrame'>
df.head()
```

```python
#Boxplot of variables
#df.boxplot(by = "Species", figsize = (10, 10))
#plt.show()

fig, axes = plt.subplots(2, 2)

ax = sns.boxplot(x="Species", y="Sepal.Length", data=df, orient='v', 
    ax=axes[0, 0])
ax = sns.boxplot(x="Species", y="Sepal.Width", data=df, orient='v', 
    ax=axes[0, 1])
ax = sns.boxplot(x="Species", y="Petal.Length", data=df, orient='v', 
    ax=axes[1, 0])
ax = sns.boxplot(x="Species", y="Petal.Width", data=df, orient='v', 
    ax=axes[1, 1])
```

```python
from pandas.plotting import parallel_coordinates
%matplotlib inline #matplotlib的圖表直接嵌入到Notebook %matplotlib inline 可以在Ipython編譯器裏直接使用，功能是可以內嵌繪圖，並且可以省略掉plt.show()這一步

parallel_coordinates(df, 'Species')

```

```python
#Pairwise correlation between attributes
sns.heatmap(df[['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width' ]].corr(), vmin = -1, vmax=1, annot=True)
plt.show()
```
